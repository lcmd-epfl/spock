```python
import warnings
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from navicat_spock.spock import run_spock_from_args
from navicat_spock.plotting2d import plot_2d
from navicat_spock.helpers import constant_data_points, augment
plt.style.use('default')
```

# Case 1: Thermocatalytic decomposition of formic acid 

```python
df1 = pd.read_csv('case 1_sabatier.csv',header=None, names=["x", "TARGET"])
df1.insert(0,'names','') # We add this because spock assumes that the first column are catalyst "names".
df1['names'] = df1.index # In this case the "names" are simply integers 1,...N
```

```python
# The figure settings are passed to spock for convenience
fig1, ax1 = plt.subplots(
    frameon=False, figsize=[4.2, 3], dpi=300, constrained_layout=True
)
# Now we run spock passing the figure and ax to get the plot in the specified format
# Verbosity is controlled by the value verb, which makes spock more verbose if set to a higher value
fig1, ax1 = run_spock_from_args(df1, verb=3, plotmode=3, fig=fig1, ax=ax1, setcbms=False, seed=42)

# We can modify the plot as we please, e.g. changing axes
#plt.ylim(-700,-200)
```
